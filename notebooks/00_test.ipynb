```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate x values
x = np.linspace(0, 2 * np.pi, 500)

# compute sine of x
y = np.sin(x)

# plot it
plt.figure(figsize=(8, 4))
plt.plot(x, y, label="sin(x)")
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.legend()
plt.show()
```
